```python
def get_categories(zip):
    # create a ChromeOptions object
    options = Options()

    # add the headless argument
    options.add_argument('--headless')

    # create a new ChromeDriver with the options
    options.add_argument('--headless')

    # specify the path to your web driver
    driver = webdriver.Chrome(options=options)

    # navigate to the login page
    driver.get('https://www.mercadona.es/')

    # enter the postal code
    postal_code = driver.find_element(By.CSS_SELECTOR, 'input[aria-label="Código postal"]')
    postal_code.send_keys(zip)

    # click the submit button
    submit_button = driver.find_element(By.CSS_SELECTOR, 'input.postal-code-form__button')
    submit_button.click()

    # wait until categories is clickable and click it.
    categorias_link = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.LINK_TEXT, "Categorías")))
    categorias_link.click()

    # Wait for the "product-cell" element to be clickable (grid of porducts)
    product_cell = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-test='product-cell']")))

    # accept cookies
    accept_button = driver.find_element(By.XPATH, "//button[contains(text(),'Aceptar todas')]")
    accept_button.click()

    # find all the category links
    category_links = driver.find_elements(By.CSS_SELECTOR, "span[class='category-menu__header']")

    ret_list = []
    for i in category_links:
        ret_list.append(i.text)
    
    driver.quit()

    return ret_list
```

```python
def get_subcategories(zip, category):
    # create a ChromeOptions object
    options = Options()

    # add the headless argument
    options.add_argument('--headless')

    # create a new ChromeDriver with the options
    options.add_argument('--headless')

    # specify the path to your web driver
    driver = webdriver.Chrome(options=options)

    # navigate to the login page
    driver.get('https://www.mercadona.es/')

    # enter the postal code
    postal_code = driver.find_element(By.CSS_SELECTOR, 'input[aria-label="Código postal"]')
    postal_code.send_keys(zip)

    # click the submit button
    submit_button = driver.find_element(By.CSS_SELECTOR, 'input.postal-code-form__button')
    submit_button.click()

    # wait until categories is clickable and click it.
    categorias_link = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.LINK_TEXT, "Categorías")))
    categorias_link.click()

    # Wait for the "product-cell" element to be clickable (grid of porducts)
    product_cell = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-test='product-cell']")))

    # accept cookies
    accept_button = driver.find_element(By.XPATH, "//button[contains(text(),'Aceptar todas')]")
    accept_button.click()

    # wait for the category to be clickable
    selected_category = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//label[text()='{category}']")))
    selected_category.click()

    # wait for the subcategories to load and save them
    subcategory_links = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".category-item__link")))

    ret_list = []
    for i in subcategory_links:
        ret_list.append(i.text)
    
    driver.quit()

    return ret_list
```

```python
def get_product_info(zip, category, subcategory):
    # create a ChromeOptions object
    options = Options()

    # add the headless argument
    options.add_argument('--headless')

    # specify the path to your web driver
    driver = webdriver.Chrome() #options=options

    # navigate to the login page
    driver.get('https://www.mercadona.es/')

    # enter the postal code
    postal_code = driver.find_element(By.CSS_SELECTOR, 'input[aria-label="Código postal"]')
    postal_code.send_keys(zip)

    # click the submit button
    submit_button = driver.find_element(By.CSS_SELECTOR, 'input.postal-code-form__button')
    submit_button.click()

    # wait until categories is clickable and click it.
    categorias_link = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.LINK_TEXT, "Categorías")))
    categorias_link.click()

    # Wait for the "product-cell" element to be clickable (grid of porducts)
    product_cell = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-test='product-cell']")))

    # accept cookies
    accept_button = driver.find_element(By.XPATH, "//button[contains(text(),'Aceptar todas')]")
    accept_button.click()

    # wait for the category to be clickable
    selected_category = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//label[text()='{category}']")))
    selected_category.click()

    # wait for the subcategories to load and save them
    selected_subcategory = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//button[text()='{subcategory}']")))
    selected_subcategory.click()

    # wait for the products to load
    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".product-cell")))

    # Find all the "product-cell" elements
    product_cells = driver.find_elements(By.CSS_SELECTOR, "div[data-test='product-cell']")

    # Get the current URL of the page
    current_url = driver.current_url

    # Click on the "product-cell" element
    product_cells[0].click()

    list_of_dicts = []

    product_count = 0
    # Iterate over the "product-cell" elements
    for i in range(len(product_cells)):
        # Scroll to the current product cell
        driver.execute_script("arguments[0].scrollIntoView();", product_cells[i])

        # Wait for the "product-cell" element to be clickable
        product_cell = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-test='product-cell']")))

        # wait for the h1 element to be present
        descripcion = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '.private-product-detail__description')))

        print(f'\rScraping "{i+1}: {product_cells[i].text[0:15]}..." product...                                                                              ', end='')
        sys.stdout.flush()

        info_prod={}

        try:
            info_prod["product"] = driver.find_element(By.CSS_SELECTOR, 'h1.title2-r.private-product-detail__description').text          #product
        except:
            info_prod["product"] = "Not available"
        
        try:
            info_prod["product_type"] = driver.find_element(By.CSS_SELECTOR, 'span.headline1-r:nth-child(1)').text                            #product_type
        except:
            info_prod["product_type"] = "Not available"
        
        try:
            info_prod["product_volume"] = driver.find_element(By.CSS_SELECTOR, 'span.headline1-r:nth-child(2)').text                            #product_volume
        except:
            info_prod["product_volume"] = "Not available"
        
        try:
            info_prod["product_price_per_unit"] = driver.find_element(By.CSS_SELECTOR, 'span.headline1-r:nth-child(3)').text
        except:
            info_prod["product_price_per_unit"] = "Not available"

        try:
            info_prod["product_price"] = float(driver.find_element(By.CSS_SELECTOR, 'p.product-price__unit-price.large-b').text.replace("€","").strip().replace(",","."))                      #product_price
        except:
            info_prod["product_price"] = "Not Available"
        
        try:
            info_prod["product_unit"] = driver.find_element(By.CSS_SELECTOR, 'p.product-price__extra-price.title1-r').text                    #product_price_unit
        except:
            info_prod["product_unit"] = "Not Available"
        
        try:
            info_prod["product_category"] = driver.find_element(By.CSS_SELECTOR, 'span.subhead1-r').text                                          #product_category
        except:
            info_prod["product_category"] = "Not Available"
        
        try:
            info_prod["product_subcategory"] = driver.find_element(By.CSS_SELECTOR, 'span.subhead1-sb').text                                         #product_subcategory
        except:
            info_prod["product_subcategory"] = "Not Available"
        
        info_prod["product_url"] = driver.current_url                                                                                    #product_url
        info_prod["product_code"] = driver.current_url.split("/")[4]
        info_prod["collected_timestamp"] = datetime.datetime.now()

        list_of_dicts.append(info_prod)

        # Send the 'esc' key
        while "categories" not in driver.current_url:
            driver.back()
            driver.find_element(By.CSS_SELECTOR, "body").send_keys(Keys.ESCAPE)

        # Click on the next "product-cell" element, if available
        if i < len(product_cells) - 1:
            next_product_cell = product_cells[i+1]
            next_product_cell.click()

            # Wait for the page to load
            WebDriverWait(driver, 10).until(EC.url_changes(current_url))

            try:
                len(driver.find_element(By.XPATH, '//button[contains(text(), "Entendido")]').text) > 0
                driver.quit()
                return "error"

            except:
                continue
        
        product_count += 1

    ret_df = pd.DataFrame(list_of_dicts)
    
    driver.quit()
    return ret_df,product_count
```
